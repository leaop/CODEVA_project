This notebook performs an exploratory data analysis (EDA) on the Students Performance dataset. The objective is to explore how various categorical factors influence studentsâ€™ scores in Math, Reading, and Writing.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (8, 5)
```

## Load Dataset

```python
df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

## Basic Information

```python
df.info()
```

## Descriptive Statistics

```python
df.describe()
```

## Check for Missing Values

```python
df.isnull().sum()
```

## Distribution of Scores

```python
for col in ['math score', 'reading score', 'writing score']:
    sns.histplot(df[col], kde=True, bins=20)
    plt.title(f'Distribution of {col}')
    plt.show()
```

## Boxplots by Gender

```python
for col in ['math score', 'reading score', 'writing score']:
    sns.boxplot(x='gender', y=col, data=df)
    plt.title(f'{col} by Gender')
    plt.show()
```

## Scatter Plots Between Scores

```python
sns.scatterplot(x='math score', y='reading score', hue='gender', data=df)
plt.title('Math Score vs Reading Score')
plt.show()

sns.scatterplot(x='reading score', y='writing score', hue='gender', data=df)
plt.title('Reading Score vs Writing Score')
plt.show()
```

## Correlation Matrix

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
